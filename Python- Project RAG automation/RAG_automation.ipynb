```python
tasks = {
    "task_id": [101, 102, 103, 104, 105, 106],
    "task_name": [
        "Site Activation",
        "Regulatory Submission",
        "Patient Recruitment",
        "Database Lock",
        "Monitoring Visit",
        "Final Study Report"
    ],
    "site_name": [
        "Chennai Site",
        "Mumbai Site",
        "Bangalore Site",
        "Hyderabad Site",
        "Delhi Site",
        "Pune Site"
    ],
    "owner": [
        "Sharon",
        "Amit",
        "Priya",
        "Rahul",
        "Neha",
        "Kiran"
    ],
    "days_overdue": [
        15,   # Overdue
        4,    # Overdue
        0,    # Due today (on time)
        -3,   # Not yet due
        -10,  # Not yet due
        1     # Slightly overdue
    ]
}
```

```python
df = pd.DataFrame(tasks)
df.head()
```

```python
def rag_status(row):
    if row['days_overdue'] > 7:
        return 'Red'
    elif row['days_overdue'] > 0:
        return 'Amber'
    else:
        return 'Green'
    
df['rag'] = df.apply(rag_status, axis=1)
df.head()
```

```python
summary = df.groupby('rag')['task_name'].count().reset_index()
summary.columns = ['Status', 'Task Count']
summary.head()
```

```python
df.to_csv('flagged_tasks_output.csv', index=False)
summary.to_csv('rag_summary.csv', index=False)

print("Done. Files exported.")
print(summary)
```
