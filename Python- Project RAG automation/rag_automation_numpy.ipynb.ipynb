```python
%pip install pandas
%pip install openpyxl
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('clinical_trial_tasks_randomised.xlsx', sheet_name='Tasks')

conditions = [
    df['days_overdue'] > 7,
    df['days_overdue'] > 0
]
choices = ['Red', 'Amber']

df['rag'] = np.select(conditions, choices, default='Green')
df.head()
```

```python
df.to_csv('flagged_tasks_output.csv', index=False)
summary = df.groupby('rag').size().reset_index(name='count')
summary.to_csv('rag_summary.csv', index=False)

print("Done. Files exported.")
print(summary)
```
