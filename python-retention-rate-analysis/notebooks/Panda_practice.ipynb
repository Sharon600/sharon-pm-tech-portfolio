# Pandas study notes

This notebook collects a few beginner-friendly pandas examples for working with site data.

## 1. Create a DataFrame

The next cell creates a DataFrame from a dictionary of site information.


```python
%pip install pandas
import pandas as pd
```

```python

site_data = {
    'site_id': [1, 2, 3, 4, 5, 6],
    'site_name': ['Chennai Central', 'Mumbai North', 'Delhi South', 'Bangalore East', 'Hyderabad West', 'Pune Metro'],
    'country': ['India', 'India', 'India', 'India', 'India', 'India'],
    'status': ['Active', 'Active', 'Pending', 'Active', 'Closed', 'Pending'],
    'enrolled_count': [22, 15, 8, 30, 12, 5],
    'days_to_activation': [45, 60, None, 38, 55, None],
}

df = pd.DataFrame(site_data)

df.head() ##shows the first few rows of a DataFrame.
```

## 2. Inspect the DataFrame

Use these commands to understand the structure of the data.


```python
print(df.shape) ##shows the number of rows and columns in the DataFrame
print(df.columns) ##shows the column names
df.info() ##shows information about the DataFrame
```

## 3. Filter rows

These examples show how to select rows based on conditions.


```python
# Pending sites and their activation days
pending_sites = df[df['status'] == 'Pending'][['site_name', 'days_to_activation']] ##filters the DataFrame to include only rows where the 'status' column is 'Pending' and selects only the 'site_name' and 'days_to_activation' columns.   
pending_sites.sort_values('days_to_activation', ascending=True) ##sorts the filtered DataFrame by the 'days_to_activation' column in ascending order.
```

## 4. Select columns

This shows how to keep only the columns you need.


```python
df[['site_name', 'enrolled_count']]
```

## 5. Sort and view top results

This helps you rank sites by a column.


```python
top_sites = df.sort_values('enrolled_count', ascending=False).head(3)
top_sites
```

## 6. Export a filtered result

This example saves a filtered DataFrame to a CSV file.


```python
high_performers = df[(df['enrolled_count'] > 15) & (df['status'] == 'Active')][['site_name', 'enrolled_count']]
high_performers.to_csv('high_performers.csv', index=False) ##saves the filtered DataFrame to a CSV file named 'high_performers.csv' without including the index column.
high_performers
```
